6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
soma_medias = df_3[['Media_provas', 'Cidade']].groupby(['Cidade']).sum() 
cont_cidades = df_3[['Media_provas', 'Cidade']].groupby(['Cidade']).count()
medias_cidades = soma_medias / cont_cidades

maior_media_cidade = medias_cidades['Media_provas'].max()
medias_cidades.query("Media_provas == @maior_media_cidade")
```

```python
colunas = ['RA', 'Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']
df_4 = pd.read_csv('alunos3.csv', sep=';', decimal=',', usecols=colunas)
```

```python
df_4['Media_provas'] = df_4[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1)
df_4['Media_provas'] = df_4['Media_provas'].round(1)

df_4
```
